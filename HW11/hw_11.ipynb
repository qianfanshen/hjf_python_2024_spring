#### 1.1.2 文本预处理
- 获取停用词表
- 对文本进行分词，保留中文，去除停用词

```python
# 获取停用词表
file_path=r'./cn_stopwords.txt'
def get_stopwords(file_path):
    '''
    获取停用词表
    Return:
        stopwords (list of strings): 停用词表
    '''
    # TODO
    with open(file_path, 'r', encoding='UTF8') as f:
        stopwords = f.read().splitlines()
        return stopwords
    
stopwords=get_stopwords(file_path)
stopwords[:10]
```

- 文本预处理
- 所有句子都变成空格分割，以方便调用sklearn的CountVectorizer
- 选1000个lable=1的句子，1000个lable=0的句子——取数据集中水果类别的前1000个和后1000个review
- 对去除的句子进行预处理，分词后的结果保存在corpus4bayes中

```python
# 文本预处理
def preprocess(text, stopwords):
    '''预处理
       Params:
            text (string): 一个源数据中的句子
            stopwords (list of strings): 停用词表
       Return:
            text (string): 预处理后的句子
    '''
    # TODO
    # 调用jieba分词
    # 仅保留中文，其余去掉
    # 去掉停用词
    text = [word for word in jieba.cut(text)]
    text = [re.sub('[^\u4e00-\u9af5]+', '', word) for word in text]
    text = [word for word in text if word not in stopwords]
    
    return text

# TODO
df_fruit = pd_all[pd_all['cat'] == '水果']
df_selected = pd.concat([df_fruit.head(1000), df_fruit.tail(1000)])
corpus = [preprocess(text, stopwords) for text in df_selected['review']]
corpus4bayes = [' '.join(sent) for sent in corpus]
corpus4bayes[:5]
```

### 1.2 在语料库级别对词汇特征进行统计:TF，IDF

- 声明TFIDF模型，默认df值从1开始，也可以设置属性过滤掉一些低频值
- 设置token_pattern=r"(?u)\b\w+\b"，包含单个汉字的词

```python
from sklearn.feature_extraction.text import CountVectorizer
# TODO
count_vec = CountVectorizer(token_pattern=r"(?u)\b\w+\b")
words_vec = count_vec.fit_transform(corpus4bayes)

print(words_vec.shape)
```

- 获取需要训练的特征`feature`
- 获取训练集和测试集 
- 用`np.random.choice`随机选取80%的数据作为训练集，剩下的作为测试集

```python
# TODO
words_dense = words_vec.toarray()
column_names = range(words_vec.shape[1])
features = pd.DataFrame(data=words_dense, columns=column_names)
all_indices = np.arange(features.shape[0])
train_size = int(len(all_indices) * 0.8)
train_index = np.random.choice(all_indices, size=train_size, replace=False)
features4test = features[~features.index.isin(train_index)]
features4train = features[features.index.isin(train_index)]
print(features4test.shape,features4train.shape)
features4test.head()
```

- 获取训练所需的标签`label`

```python
# TODO
labels = df_selected['label']
label4test = labels[~features.index.isin(train_index)]
label4train = labels[features.index.isin(train_index)]
print(label4test.shape,label4train.shape)
label4test.head()
```

- 构建数据集的函数`construct_features`
- 构建训练集与测试集
- 进行贝叶斯分类

```python
%pip install nltk
```

```python
import nltk
# 获取特征名
feature_names=count_vec.get_feature_names_out()
# 构建特征
def construct_features(features,labels,feature_names):
    # TODO
    feature_list = []
    for i in range(features.shape[0]) :
        text_feature = {}
        for j in range (features.shape[1]) :
            text_feature[feature_names[j]] = features.iloc[i, j] 
        feature_list.append((text_feature, labels[i]))
    return feature_list

# TODO
# 构建训练集和测试集
train_dataset = construct_features(features4train, label4train.tolist() , feature_names) 
test_dataset = construct_features (features4test, label4test.tolist(), feature_names)
classifier = nltk.NaiveBayesClassifier.train(train_dataset)
print(nltk.classify.accuracy(classifier, test_dataset))
classifier.show_most_informative_features(5)
```

- 将切分后的数据作为新的一列`word`加入到数据集中，按照课件中的操作，将切分的词用空格连接为string

```python
# TODO
def split_text(text):
    words = jieba.cut(text)
    words = [word for word in words if word not in stopwords]
    return ' '.join(words)
pd_data['words'] = pd_data['review'].apply(split_text)
pd_data.head()
```

- 设计通过情感词典进行的句子情感特征提取，设计初步方案

```python
def sentiment_analysis(df,sentiment_words,strength_words):
    '''
    sentiment_score:该句的总情感强度得分
    positive_score:该句的正向情感强度得分
    negative_score:该句的负向情感强度得分
    '''
    sentiment_score=0
    positive_score=0
    negative_score=0
    # 遍历当前句的每个词
    # TODO
    words = df['words'].split()
    for w in words:
        if w in sentiment_words.keys():
            if sentiment_words[w] == 1:
                sentiment_score += strength_words[w]
                positive_score += strength_words[w]
            if sentiment_words[w] == 0:
                sentiment_score -= strength_words[w]
                negative_score += strength_words[w]
    
    return sentiment_score,positive_score,negative_score            
```

- 构造新的特征（三个score）`new_features`
- 获取数据的标签`labels`

```python
# TODO
results = pd_data.apply(lambda row: sentiment_analysis(row, sentiment_words, strength_words), axis=1)
new_features = pd.DataFrame(list(results), columns=['sentiment_score', 'positive_score', 'negative_score'])
labels = pd_data['label']
new_features,labels
```

常规的向量化-降维-贝叶斯分类训练

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# TODO
# 进行训练集和测试集的划分，test_size=0.1
X_train1, X_test1, y_train1, y_test1 = train_test_split(new_features, labels, test_size=0.1, random_state=True)
# 进行贝叶斯模型的初始化和训练
gnb = GaussianNB()
gnb.fit(X_train1, y_train1)

# # 在测试集上进行预测
y_pred1 = gnb.predict(X_test1)
# 计算预测的效果，使用f1_score
f1 = f1_score(y_pred1, y_test1, average='binary')
print('F1 score:',f1)
```

#### 1.3.1 情感词典优化

```python
# 情感分析函数
def my_sentiment_analysis(df,sentiment_words,strength_words):
    '''
    sentiment_score:该句的总情感强度得分
    positive_score:该句的正向情感强度得分
    negative_score:该句的负向情感强度得分
    sentiment:该句的情感极性，1表示正向，0表示负向
    positive_count:正向词占比
    negative_count:负向词占比
    '''
    positive_count=0
    negative_count=0
    positive_score=0
    negative_score=0
    
    length=len(df)
    
    # TODO
    words = df['words'].split()
    for w in words:
        if w in sentiment_words.keys():
            if sentiment_words[w] == 1:
                positive_count+=1
                positive_score += strength_words[w]
            if sentiment_words[w] == 2:
                negative_count+=1
                negative_score += strength_words[w]
    positive_count = positive_count / length
    negative_count = negative_count / length
    sentiment_score = positive_score - negative_score
    sentiment = 1 if sentiment_score > 0 else 0
    
    return sentiment_score,positive_score,negative_score,sentiment,positive_count,negative_count
```

- 构造新的特征`new_features1`

```python
# TODO
results = pd_data.apply(lambda row: my_sentiment_analysis(row, sentiment_words, strength_words), axis=1)
new_features1 = pd.DataFrame(list(results), columns=['sentiment_score', 'positive_score', 'negative_score', 'sentiment', 'positive_cnt', 'negative_cnt'])
new_features1
```

使用新的特征进行常规的向量化-降维-贝叶斯分类训练

```python
# TODO
# 进行训练集和测试集的划分，test_size=0.1
X_train2, X_test2, y_train2, y_test2 = train_test_split(new_features1, labels, test_size=0.1, random_state=True)
# 进行贝叶斯模型的初始化和训练
gnb = GaussianNB()
gnb.fit(X_train2, y_train2)

# # 在测试集上进行预测
y_pred2 = gnb.predict(X_test2)
# 计算预测的效果，使用f1_score
f1 = f1_score(y_pred2, y_test2, average='binary')

print('F1 score:',f1)
```

### 2.1.3 添加节点属性
对每一个节点添加weight属性，其值为节点的度数，并绘制带标签的图片，要求节点的大小和weight的大小成正比

```python
#TODO:
for node in G.nodes():
    G.nodes[node]['weight'] = G.degree[node]
node_weights = [G.nodes[node]['weight'] * 300 for node in G.nodes()]
nx.draw(G, with_labels=True, node_size = node_weights)
```

#### 2.1.4 点中心度计算

```python
#TODO:
for node in G.nodes():
    G.nodes[node]['centrality'] = nx.degree_centrality(G)[node]
for node in G.nodes():
    weight = G.nodes[node].get('weight', 0)
    centrality = G.nodes[node]['centrality']
    print("Node:{} Attributes: {{'weight': {}, 'centrality': {}}}".format(node, weight, centrality))
```
